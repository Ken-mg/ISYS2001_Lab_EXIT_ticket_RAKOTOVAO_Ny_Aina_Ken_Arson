## Part 4: Building Your Finance Tracker Foundation (15 minutes)

### Start Your Semester Project

Create a new notebook called "Personal_Finance_Tracker" and build the foundation:

````python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# Expense entry
print("\n--- Expense Entry ---")

expense_amount = float(input("Enter an expense amountper months: $"))
expense_description = input("What was this expense for? ")

saving_goal = float(input("Enter your saving goal for this month: $"))

# Calculate expense percentage
expense_ratio = (expense_amount / monthly_income) * 100

# Categorise the expense
if expense_ratio < 5:
    category = "Small expense"
    advice = "Good spending control."

elif expense_ratio < 15:
    category = "Moderate expense"
    advice = "Reasonable expense."

elif expense_ratio < 30:
    category = "Large expense"
    advice = "Be careful with spending."

else:
    category = "Very large expense"
    advice = "Warning: This expense is very high!"

# Calculate remaining money
remaining_money = monthly_income - expense_amount

# Check savings goal
if remaining_money >= saving_goal:
    saving_message = "You can still reach your saving goal this month."
else:
    saving_message = "You may need to reduce spending to reach your saving goal."

# Results
print("\n--- Expense Analysis ---")
print(f"Expense description: {expense_description}")
print(f"Expense amount: ${expense_amount:.2f}")
print(f"This represents {expense_ratio:.1f}% of your monthly income.")
print(f"Category: {category}")
print(advice)

print("\n--- Budget Status ---")
print(f"Remaining money after expense: ${remaining_money:.2f}")
print(saving_message)
````
