### Paste and Test Your AI-Generated Budget Tracker

```python
3# AI-GENERATED BUDGET TRACKER CODE
# [# Ask for monthly income
income = float(input("Enter your monthly income: "))

# Ask for up to 3 expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\nTotal Expenses:", total_expenses)
print("Remaining Budget:", remaining_budget)

# Warning if budget is negative
if remaining_budget < 0:
    print("⚠ Warning: Your budget is negative! You are overspending.")
else:
    print("✅ Good job! You are within your budget.")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: [Show percentage of income spent ]"
enhancement_prompt = """
[i want you to create a programm that : Allows users to input their monthly income Allows users to input up to 3 expenses Calculates remaining budget Shows a warning if budget is negative . However , had this new feature in the programm : Show percentage of income spent]
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
# [# Ask for monthly income
income = float(input("Enter your monthly income: "))

# Initialize total expenses
total_expenses = 0

# Allow user to enter up to 3 expenses
for i in range(1, 4):
    expense = float(input(f"Enter expense {i}: "))
    total_expenses += expense

# Calculate remaining budget
remaining_budget = income - total_expenses

# Calculate percentage of income spent
percentage_spent = (total_expenses / income) * 100

# Display results
print("\n----- Budget Summary -----")
print(f"Total Expenses: {total_expenses:.2f}")
print(f"Remaining Budget: {remaining_budget:.2f}")
print(f"Percentage of Income Spent: {percentage_spent:.2f}%")

# Warning if budget is negative
if remaining_budget < 0:
    print("⚠ Warning: Your budget is negative! You are overspending.")
else:
    print("✅ Good job! You are within your budget.")
```
