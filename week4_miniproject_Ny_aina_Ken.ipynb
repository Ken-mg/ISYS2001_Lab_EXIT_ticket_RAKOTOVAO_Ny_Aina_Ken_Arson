## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
from ast import Try
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts
descriptions = []  # Store what each expense was for
total_spent = 0


print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")



# Your task: Complete the multi-expense entry system
# Use sentinel value pattern to collect multiple expenses
# Validate each expense amount (positive numbers only)
# Store both amount and description
# Show running total and budget percentage after each entry

while True:
    expense_input = input("\nExpense amount (or 'done'): $")


    if expense_input.lower() == 'done':
        break

    try:
        #User amount price
        amount = float(expense_input)
        #Amound and description list
        if amount <= 0:
            print("Please enter a positive number!")
        else:

            item = input("What was this expense for? ")

            expenses.append(amount)
            descriptions.append(item)
            total_spent = total_spent + amount


            percent = (total_spent / monthly_budget) * 100
            print(f"Added: {item} (${amount}) added to your list")
            print(f"Total so far: ${total_spent:.2f} ({percent:.1f}% of budget)")

    except ValueError:
        print("Invalid input! Please enter a number.")

print(f"\n Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
print(f"\n Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Expense Analysis Section

print("Basic Expense Analysis")
#1 User name ; User monthly budget and user expense
User_name = input("Enter Your name :")

monthly_budget = 0

while monthly_budget < 100 or monthly_budget > 10000:
        try:
          budget_input = input("Enter monthly budget ($100-$10,000): $")
          monthly_budget = float(budget_input)

          if monthly_budget < 100:
            print("Budget too low! Minimum $100")
          elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
        except ValueError:
           print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")



if len(expenses) > 0:
    print("\n📊 Expense Analysis:")

    # Basic statistics using loop patterns from workshops
    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")
    print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Your task: Find highest and lowest expenses
    # Use the min/max pattern from workshops
    highest_expense = 0
    lowest_expense = 0

    # 1.Min/ Max code
    for amount in expenses:
        if amount > highest_expense:
            highest_expense = amount

        if amount < lowest_expense:
            lowest_expense = amount


    print(f"\n💰 Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # 2 Categorisation of expenses

    large_count = 0    # > 20% of budget
    medium_count = 0   # 5-20% of budget
    small_count = 0    # < 5% of budget

    for amount in expenses:
      item_percentage = (amount / monthly_budget) * 100
    if item_percentage > 20:
        large_count = large_count + 1
    elif item_percentage >= 5:
        medium_count = medium_count +1
    else :
      small_count = small_count+1



    print(f"\n📊 Expense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")


    print("\n💡 Smart Feedback:")

    # 3. Feedback provided
    if total_spent > monthly_budget:
      print ( " Warning !You have exceeded your budget , you need to cut costs" )
    elif total_spent >= monthly_budget *0.9:
      print("Careful , you are almost out of money there . Think it twice")
    else :
      print(" Great Job ! Stay with that consistance ")

    if large_count >=2 :
      print ( "Maybe you should try to spread more your expenses")
    if small_count > 5:
      print("These Tiny costs add up fast ")
else:
    print("\n📝 No expenses entered - try adding some transactions!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []


# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

# Your task: Collect goal information using for loop
# For each goal (1 to num_goals):
#   - Get goal name
#   - Get target amount (with validation)
#   - Get current amount (with validation)
#   - Store in the three lists

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")


    name = input("What is the name for this goal? ")

    target = 0
    while target <= 0:
        try:
            target = float(input(f"Target amount for {name}: $"))
            if target <= 0:
                print("Please enter a positive number!")
        except ValueError:
            print("Please enter a valid amount!")

    current = -1
    while current < 0:
        try:
            current = float(input(f"Current amount saved for {name}: $"))
            if current < 0:
                print("Please enter a positive number!")
        except ValueError:
            print("Please enter a valid amount!")


    goal_names.append(name)
    goal_targets.append(target)
    goal_current.append(current)


print(f"\n✅ Saved {len(goal_names)} goals!")

print(f"\n✅ Saved {len(goal_names)} goals!")
```

```python
# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    # Your task: Analyze each goal
    # For each goal, calculate:
    #   - Progress percentage
    #   - Amount remaining
    #   - Months to completion (if saving $100/month)

    total_target = 0
    total_current = 0

    for i in range(len(goal_names)):
        # Your progress calculation here

            total_target = 0
            total_current = 0


    for name, target, current in zip(goal_names, goal_targets, goal_current):


        percent = (current / target) * 100 if target > 0 else 0
        remaining = target - current

        print(f"\n--- {name} ---")
        print(f"Progress: {percent:.1f}%")
        print(f"Remaining: ${remaining:.2f}")

        if remaining > 0:
            months = remaining / 100  # Estimation
            print(f"Time left (at $100/mo): {months:.1f} months")
        else:
            print("🎉 Goal achieved! Well done!")

        total_target += target
        total_current += current
        total_target += goal_targets[i]
        total_current += goal_current[i]

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")


    lowest_percent = 101
    attention_goal = ""


    for i in range(len(goal_names)):
        # 1 Calculating the percentage for the current goal
        current_prog = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0

        if current_prog < lowest_percent:
            lowest_percent = current_prog
            attention_goal = goal_names[i]

    # 2. Displaying personalized recommendations
    print(f"\n💡 Smart Advice:")
    if lowest_percent < 100:
        print(f"Your goal '{attention_goal}' needs the most attention (only {lowest_percent:.1f}% reached).")
        print(f"Keep pushing! Small savings every day will get you there! ")
    else:
        print("Amazing! All your savings goals are 100% complete! ")

else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python

name = "User"
monthly_budget = 200
expenses = [50]
expense_summary = []
goal_names = ["Food", "Life", "Video game"]
goal_targets = [25.0, 60.0, 15.0]

goal_current = [20.0, 55.0, 10.0]

print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print("💰 " + name + "'s Finance Tracker")


    if len(expenses) > 0:
        total_now = sum(expenses)
        print("📊 Current: " + str(len(expenses)) + " expenses, $" + str(round(total_now, 2)) + " spent")

    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        print("\n--- Expense Summary ---")
        if len(expenses) == 0:
            print("No expenses yet! Try adding some with option 3.")
        else:
            for i in range(len(expenses)):
                price = expenses[i]
                item = expense_summary[i]
                count = i + 1
                print(str(count) + ". $" + str(price) + " - " + item)

            total_cost = sum(expenses)
            print("-" * 20)
            print("TOTAL: $" + str(total_cost))

    elif choice == "2":
        print("\n--- Goals Progress ---")

        if len(goal_names) == 0:
            print("No goals set yet! Please add some first.")
        else:
            for i in range(len(goal_names)):
                name_goal = goal_names[i]
                current_val = goal_current[i]
                target_val = goal_targets[i]


                percent = (current_val / target_val) * 100

                print(name_goal + ": $" + str(current_val) + "/$" + str(target_val) + " (" + str(round(percent, 1)) + "% complete)")

    elif choice == "3":
        print("\n--- Add New Expense ---")
        new_item = input("What did you buy? ")

        valid_price = False
        while valid_price == False:
            price_input = input("How much did it cost? $")
            try:
                new_price = float(price_input)
                if new_price > 0:
                    valid_price = True
                else:
                    print("Please enter a price higher than 0.")
            except:
                print("Error: Please enter a number (example: 15.50)")

        expenses.append(new_price)
        expense_summary.append(new_item)

        print("\n✅ Success!")
        print("Added $" + str(new_price) + " for " + new_item + " to your list.")

        current_total = sum(expenses)
        print("Your total expenses are now: $" + str(round(current_total, 2)))

    elif choice == "4":
        print("\n--- Budget Check ---")
        total_spent = sum(expenses)
        money_left = monthly_budget - total_spent

        if monthly_budget > 0:
            percent_used = (total_spent / monthly_budget) * 100
        else:
            percent_used = 0

        daily_budget = money_left / 30

        print("Budget Limit: $" + str(monthly_budget))
        print("Total Spent:  $" + str(total_spent))
        print("Money Left:   $" + str(money_left))
        print("Status: " + str(round(percent_used, 1)) + "% of budget used")

        if money_left > 0:
            print("You can spend $" + str(round(daily_budget, 2)) + " per day for the rest of the month.")
        else:
            print("You have no money left for the month!")

        print("\n--- Advice ---")
        if percent_used >= 100:
            print("🚨 STOP! You are over budget. Cut all non-essential spending!")
        elif percent_used >= 80:
            print("⚠️ Careful! You have used most of your budget. Be thrifty.")
        elif percent_used >= 50:
            print("👍 You are halfway through. Watch your upcoming bills.")
        else:
            print("✅ Great job! You are managing your money very well.")

    elif choice == "5":
        running = False
        total_final = sum(expenses)
        number_of_expenses = len(expenses)

        print("\nGoodbye " + name + "!")
        print("Final summary: " + str(number_of_expenses) + " expenses recorded.")
        print("Total amount spent: $" + str(round(total_final, 2)))
        print("Keep tracking your finances! See you next time. 👋")

    else:
        print("Invalid choice! Please enter a number between 1 and 5.")

print("\n=== Finance Tracker Session Complete ===")
```
